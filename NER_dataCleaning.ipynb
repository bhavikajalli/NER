```python
import pandas as pd
```

```python
#names = pd.read_csv('person_names.csv',names = ['String','Cleaned_string','Honorific','First_Name','Middle_Name','Last_name','Suffix','Country','Profession'])
names = pd.read_csv('names_gender.csv',encoding = "ISO-8859-1")
```

```python
names = names[['string','gender','honorific','first_name','middle_name','last_name','suffix']]
names.head(20)
```

```python
for i in range(len(names)):
    parts = names['string'][i].strip().split(" ")
    #print(parts)
    if parts[0] == 'Sir':
        names['honorific'][i] = parts[0]
        print(parts[0]) 
    suff = names['string'][i].strip().split(",") 
    suf = suff[-1].split(" ")[0]
    if suf == 'Jr.' or suf == 'Sr.':
        names['suffix'][i] = suf
        print(names['suffix'][i])
        print(i)
    if parts[-1] == 'Jr.' or parts[-1] == 'Sr.':
        names['suffix'][i] = parts[-1]
        print(names['suffix'][i])
        print(i)
    if len(parts) > 2:
        try:
            end = parts.index(names['last_name'][i])
        except ValueError:
            pass
        #print(parts)
        names['middle_name'][i] = " ".join(part for part in parts[1:end])
        #print(names['middle_name'][i])
       
    if i % 10000 == 0:
        print(i)
    
```

```python
names.to_csv('names_cleaned.csv', sep='\t', encoding='utf-8')
```

```python
occupation = pd.read_csv('national_M2017_dl.csv',encoding = "ISO-8859-1")
```

```python
occupation.head()
```

```python
import pickle
```

```python
occupations = []
for row in occupation['OCC_TITLE']:
    occs = row.strip().split(",")
    if ' and' in occs:
        occs.remove('and')
    for occ in occs:
        occ_and = occ.split("and")
        for occ_and_term in occ_and:
            occupations.append(occ_and_term.strip())
        
occupations = list(set(occupations))
```

```python
occupations.remove('')
```

```python
occupations
```

```python
with open('occupations.pkl','wb') as pickleFile:
    pickle.dump(occupations,pickleFile)
```
